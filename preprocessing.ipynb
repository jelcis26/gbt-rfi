```python
import pandas as pd
import pandasql as ps #SQL on Pandas Dataframe
```

```python
carmen_df = pd.read_csv('/home/cgchoza/galaxies/complete_cadences_catalog.csv')
carmen_df
```

```python
pd.set_option('display.max_colwidth', None)
print(carmen_df['.h5 path'].iloc[:6])  # Print first 6 rows of the column
```

```python
# Extract two numbers after 'guppi_' using regex
carmen_df[['MJD', 'seconds']] = carmen_df['.h5 path'].str.extract(r'guppi_(\d+)_(\d+)')

# Optional: convert to integers
carmen_df['MJD'] = carmen_df['MJD'].astype(int)
carmen_df['seconds'] = carmen_df['seconds'].astype(int)
carmen_df
```

```python
sort_query = '''
SELECT *
FROM carmen_df
ORDER BY
  "Cadence ID"   ASC,
  "MJD"          ASC,
  "seconds"      ASC
'''
sorted_df = ps.sqldf(sort_query, locals()).reset_index(drop=True)
sorted_df
```

```python
sort_query = '''
SELECT "Band", COUNT("Band") AS count
FROM carmen_df
GROUP BY "Band"
'''
Band_df = ps.sqldf(sort_query, locals()).reset_index(drop=True)
Band_df
```

```python
sort_query = '''
SELECT "Cadence ID", COUNT("Cadence ID") AS count
FROM carmen_df
GROUP BY "Cadence ID"
ORDER BY count DESC
'''
Cadence_df = ps.sqldf(sort_query, locals()).reset_index(drop=True)
Cadence_df
#Cadence_df.head(30)
```

```python
sort_query = '''
SELECT "Cadence ID", "Band", COUNT("Band") AS count
FROM carmen_df
GROUP BY "Cadence ID", "Band"
ORDER BY count DESC
'''
Cadence_Band_df = ps.sqldf(sort_query, locals()).reset_index(drop=True)
Cadence_Band_df
#Cadence_Band_df.head(30)
```

```python
sort_query = '''
SELECT "Cadence ID", "Band", "Frequency", COUNT("Frequency") AS count
FROM carmen_df
GROUP BY "Cadence ID", "Band", "Frequency"
ORDER by count DESC
'''
Cadence_Band_Frequency_df = ps.sqldf(sort_query, locals()).reset_index(drop=True)
Cadence_Band_Frequency_df
#Cadence_Band_Frequency_df.head(20)
```

```python
sorted_df_411390 = sorted_df[sorted_df['Cadence ID'] == 411390]
carmen_df_411390 = carmen_df[carmen_df['Cadence ID'] == 411390]

def print_non_matching_rows(df1: pd.DataFrame, df2: pd.DataFrame) -> None:
    # Create a tuple representation of each row to allow set comparisons
    df1_tuples = df1.apply(tuple, axis=1)
    df2_tuples = df2.apply(tuple, axis=1)

    # Get rows unique to df1 and to df2
    df1_unique = df1[~df1_tuples.isin(df2_tuples)]
    df2_unique = df2[~df2_tuples.isin(df1_tuples)]

    print("Rows in df1 not in df2:")
    print(df1_unique)
    print("\nRows in df2 not in df1:")
    print(df2_unique)
print_non_matching_rows(sorted_df, carmen_df)
```
